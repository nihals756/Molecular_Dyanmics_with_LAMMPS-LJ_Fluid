# RDF

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Snapshots

Total RDF

```python
df=pd.read_excel('RDF.xlsx')

df.plot(kind='line',x='r0',y='rdf0',label='t=0',color='red',figsize=(15,5))
plt.title('RDF vs r')
plt.xlabel('r')
plt.ylabel('RDF')
plt.xticks(ticks=range(1, int(df[['r0', 'r1', 'r2', 'r3']].max().max()) + 1))
plt.legend()

df.plot(kind='line',x='r1',y='rdf1',label='t=100000',color='blue',figsize=(15,5))
plt.title('RDF vs r')
plt.xlabel('r')
plt.ylabel('RDF')
plt.xticks(ticks=range(1, int(df[['r0', 'r1', 'r2', 'r3']].max().max()) + 1))
plt.legend()

df.plot(kind='line',x='r2',y='rdf2',label='t=200000',color='green',figsize=(15,5))
plt.title('RDF vs r')
plt.xlabel('r')
plt.ylabel('RDF')
plt.xticks(ticks=range(1, int(df[['r0', 'r1', 'r2', 'r3']].max().max()) + 1))
plt.legend()

df.plot(kind='line',x='r3',y='rdf3',label='t=300000',color='purple',figsize=(15,5))
plt.title('RDF vs r')
plt.xlabel('r')
plt.ylabel('RDF')
plt.xticks(ticks=range(1, int(df[['r0', 'r1', 'r2', 'r3']].max().max()) + 1))
plt.legend()
```
