```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('pcf.xlsx')
df.dropna(inplace=True)
df_av=df.groupby('r',as_index=False).mean()

df11=pd.read_excel('pcf11.xlsx')
df11.dropna(inplace=True)
df11_av=df11.groupby('r',as_index=False).mean()

df22=pd.read_excel('pcf22.xlsx')
df22.dropna(inplace=True)
df22_av=df22.groupby('r',as_index=False).mean()

df12=pd.read_excel('pcf12.xlsx')
df12.dropna(inplace=True)
df12_av=df12.groupby('r',as_index=False).mean()
```

```python
df.shape
```

```python
df_av.shape
```

```python
fig, ax= plt.subplots(figsize=(15,5))
df_av.plot(x='r',y='rdf',kind='line',ax=ax,label='Tot')
df11_av.plot(x='r',y='rdf',kind='line',ax=ax,label='1-1')
df22_av.plot(x='r',y='rdf',kind='line',ax=ax,label='2-2')
df12_av.plot(x='r',y='rdf',kind='line',ax=ax,label='1-2')
plt.axhline(y=1, color='k', linestyle='--', linewidth=2)
plt.legend()
plt.xlabel('r')
plt.ylabel('rdf')
plt.show()
```

```python
max_rdf_ind=df_av['rdf'].idxmax()
print('Maxima of',df_av['rdf'].max(),' at r=',df_av.loc[max_rdf_ind,'r'])
```
